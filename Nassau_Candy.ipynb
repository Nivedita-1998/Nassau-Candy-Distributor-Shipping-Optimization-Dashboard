# **DATA CLEANING & PROCESSING**

* **Date Format Standardization**









```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Order Date']
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

df['Ship Date']
```

```python
df.dtypes[['Order Date','Ship Date' ]]
```

* **STANDARDIZE LABELS - Standardize Column Names**

```python
df.columns = df.columns.str.strip()

df.columns
```

 * **Invalid Records Removal**





```python
# Remove rows which are zero.
df = df.dropna()
```

* **Removing Duplicates**

```python
duplicates = df.duplicated()
df = df.drop_duplicates()
print(f"Duplicate Records Found: {duplicates.sum()}")
df.drop_duplicates()
```

* **HANDLE MISSING VALUES**

```python
df['Sales'] = df['Sales'].fillna(0)
df['Units'] = df['Units'].fillna(0)
df['Gross Profit'] = df['Gross Profit'].fillna(0)
```

```python
df.head()
```

# **FEATURE ENGINEERING**

 * **CALCULATE SHIPPING LEAD TIME (PRIMARY TARGET VARIABLE)**








```python
df['Lead Time'] = (df['Ship Date'] - df['Order Date']).dt.days

# Remove negative or zero lead times
df = df[df['Lead Time'] >= 0]
df['Lead Time']
```

```python
df['Lead Time'].describe().round(2)
```

```python
#Delay Flag (mean + 1 std)
DELAY_THRESHOLD    = df['Lead Time'].mean() + df['Lead Time'].std()
df['Delayed']      = df['Lead Time'] > DELAY_THRESHOLD
df['Delayed']
```

```python
print(f"Delay threshold: {DELAY_THRESHOLD:.1f} days")
print(f"Delayed shipments: {df['Delayed'].sum():,} ({df['Delayed'].mean()*100:.1f}%)")
```

* **PROFIT MARGIN**

```python
total_profit = df['Gross Profit'].sum()
df['Profit_Margin'] = df['Gross Profit'] / df['Sales']
df['Profit_Margin %'] = df['Gross Profit'] / df['Sales'] * 100

df['Profit_Margin %']
```

```python
total_sales = df['Sales'].sum()
print(f"Profit Margin: {(total_profit / total_sales * 100):.1f}%")
```

* **PROFIT CONTRIBUTION %**


```python
total_profit = df['Gross Profit'].sum()
df['Profit Contribution %'] = (df['Gross Profit'] / total_profit) * 100
df['Profit Contribution %']
```

* **REVENUE PER UNIT**

```python
df['Revenue_Per_Unit'] = df['Sales'] / df['Units']
df['Revenue_Per_Unit']
```

* **REVENUE CONTRIBUTION**

```python
total_sales = df['Sales'].sum()
df['Revenue Contribution %'] = (df['Sales'] / total_sales) * 100
df['Revenue Contribution %']
```

* **ORDER MONTH AND DAY OF WEEK**

```python
df['Order_Year'] = df['Order Date'].dt.year
df['Order_Month'] = df['Order Date'].dt.month
df['Order_Day Of Week'] = df['Order Date'].dt.dayofweek
```

* **SHIP MONTH AND DAY OF WEEK**

```python
df['Ship_Year'] = df['Ship Date'].dt.year
df['Ship_Month'] = df['Ship Date'].dt.month
df['Ship_Day Of Week'] = df['Ship Date'].dt.dayofweek
```

 * **AGGREGATE ROUTE LEVEL METRICS**

```python
# Aggrigation by route Region

route_region_summary = df.groupby('Route_Region').agg(
    Total_Shipments=('Route_Region', 'count'),
    Avg_Lead_Time=('Lead Time', 'mean'),
    Lead_Time_Variability=('Lead Time', 'std')
).reset_index()
```

```python
# Rank by Region

route_region_ranked = route_region_summary.sort_values(by='Avg_Lead_Time')
```

```python
# Aggregation by route state

route_state_summary = df.groupby('Route_State').agg(
    Total_Shipments=('Route_State', 'count'),
    Avg_Lead_Time=('Lead Time', 'mean'),
    Lead_Time_Variability=('Lead Time', 'std')
).reset_index()
route_state_summary.head(5)
```

```python
# Rank by State

route_state_ranked = route_state_summary.sort_values(by='Avg_Lead_Time')
```

```python
# Top 10 fastest route
top_10_routes = route_state_ranked.head(10)
top_10_routes
```

```python
# Bottom 10 slowest
bottom_10_routes = route_state_ranked.tail(10)
bottom_10_routes
```

```python
# High average lead time

avg_lead_time = df.groupby('State/Province')['Lead Time'].mean().sort_values(ascending=False)

avg_lead_time.head(10)
```

```python
# Product-Level Profitability Analysis
product_analysis = df.groupby('Product Name').agg(
                  Total_Sales=('Sales','sum'),
                  Total_Profit=('Gross Profit','sum'),
                  Total_Units=('Units','sum')
                  ).reset_index()

product_analysis['Margin %'] = (product_analysis['Total_Profit'] / product_analysis['Total_Sales']) * 100
product_analysis['Profit per Unit'] = product_analysis['Total_Profit'] / product_analysis['Total_Units']

product_analysis.head()
```

```python
# Congation-Prone Region
congestion = df.groupby('State/Province').agg({
    'Lead Time': ['mean', 'std'],
    'Delayed': 'mean'
})
congestion.columns = ['Avg_Lead_Time', 'Std_Dev', 'Delay_Frequency']
```

```python
# Sort by delay frequency
congestion = congestion.sort_values(by='Delay_Frequency', ascending=False)

congestion.head(10)
```

```python
# Average Lead Time by Ship Mode
ship_eff = df.groupby('Ship Mode')['Lead Time'].mean().sort_values()

ship_eff
```

```python
ship_mode_perf = df.groupby('Ship Mode').agg({
    'Lead Time': ['mean', 'std'],
    'Order Date': 'count'
})

ship_mode_perf.columns = ['Avg_Lead_Time', 'Std_Dev', 'Volume']

ship_mode_perf.sort_values(by='Avg_Lead_Time')
```

```python
# Division-Level Performance Analysis

division_analysis = df.groupby('Division').agg(
    Sales=('Sales','sum'),
    Profit=('Gross Profit','sum'),
    Units=('Units','sum'),
    Avg_Lead_Time=('Lead Time','mean')
).reset_index()

division_analysis['Margin %'] = (division_analysis['Profit'] / division_analysis['Sales']) * 100
division_analysis['Delay_Rate'] = df.assign(Delayed=df['Lead Time'] > delay_threshold).groupby('Division')['Delayed'].mean() * 100
division_analysis
```

* **BOTTLENECK ANALYSIS**

```python
# Bottleneck States
volume = df.groupby('State/Province').size()
performance = df.groupby('State/Province')['Lead Time'].mean()
bottleneck_df = pd.DataFrame({
    'Volume': volume,
    'Avg_Lead_Time': performance
}).sort_values(by=['Volume', 'Avg_Lead_Time'], ascending=False).head(10)

print("=== Top Bottleneck States ===")
print(bottleneck_df)
```

```python
# State-level performance ───────────────────────────────────────────────────
state_perf = df.groupby('State/Province').agg(
    Avg_Lead   = ('Lead Time', 'mean'),
    Volume     = ('Row ID',    'count'),
    Delay_Rate = ('Delayed',   'mean'),
).reset_index().sort_values('Avg_Lead', ascending=False)

# Top 15 worst states
worst_states = state_perf.head(15)
fig, ax = plt.subplots(figsize=(13, 6))
colors = plt.cm.Reds(np.linspace(0.4, 0.9, 15))
ax.barh(worst_states['State/Province'], worst_states['Avg_Lead'], color=colors, edgecolor='white')
ax.set_title('Top 15 States with Highest Avg Lead Time (Bottlenecks)', fontsize=13, fontweight='bold')
ax.set_xlabel('Avg Lead Time (days)', fontsize=13, fontweight='bold')
ax.set_ylabel('State/Province', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
